```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_excel(r"C:\Users\sunda\OneDrive\Documents\GENAI\houseprice.xlsx")
df
```

```python
plt.xlabel("Area")
plt.ylabel("Prices")
plt.scatter(df.area,df.price,color="blue",marker=".")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
#reg.fit() is used for training the dataset it accepts two argument where 1st argument should be 2d array.
```

```python
reg.predict([[3400]])
```

```python
d=pd.read_excel(r"C:\Users\sunda\OneDrive\Documents\test_data_house_price.xlsx")
d
```

```python
price=reg.predict(d[['area']])
#creating new col in data frame.
d['price']=price
#exporting data frame as csv file
d.to_csv("predictions.csv")
d
```
